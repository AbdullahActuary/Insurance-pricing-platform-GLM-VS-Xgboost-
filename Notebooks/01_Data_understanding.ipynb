## 5. Missing Value Assessment

Missing values can affect model performance and statistical analysis. This section evaluates the completeness of each dataset by calculating the number and percentage of missing values for every variable.

```python
# Missing values - Frequency dataset

missing_freq = (
    frequency
    .isnull()
    .sum()
    .to_frame(name="Missing Values")
)

missing_freq["Percentage (%)"] = (
    missing_freq["Missing Values"]
    / len(frequency)
    * 100
).round(2)

missing_freq.sort_values(
    by="Missing Values",
    ascending=False
)
```

```python
# Missing values - Severity dataset

missing_sev = (
    severity
    .isnull()
    .sum()
    .to_frame(name="Missing Values")
)

missing_sev["Percentage (%)"] = (
    missing_sev["Missing Values"]
    / len(severity)
    * 100
).round(2)

missing_sev.sort_values(
    by="Missing Values",
    ascending=False
)
```

```python
print("Frequency Dataset Duplicates :", frequency.duplicated().sum())
print("Severity Dataset Duplicates  :", severity.duplicated().sum())
```
